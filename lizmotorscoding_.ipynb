```python
## snippet to extract information about canoo from internet



def extract_canoo_information(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all paragraph elements containing the word 'canoo'
        canoo_paragraphs = soup.find_all('p', string=lambda text: 'canoo' in text.lower() if text else False)

        # Extract and print information from each paragraph
        print(f'Information about "canoo":')
        for paragraph in canoo_paragraphs:
            print(paragraph.text)

    else:
        print(f'Error: Unable to fetch data from {url}. Status code: {response.status_code}')


extract_canoo_information(url)




```

```python

import pandas as pd

def retrieve_tables_to_dataframe(url):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all tables in the HTML
        tables = soup.find_all('table')

        # List to store DataFrames for each table
        dfs = []

        # Iterate through each table and convert it to a DataFrame
        for i, table in enumerate(tables):
            # Extract data from the table
            table_data = []
            header_row = None

            for row in table.find_all('tr'):
                row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]

                # If header_row is not set, assume the current row is the header
                if header_row is None:
                    header_row = row_data
                else:
                    # Check if the current row has the same number of columns as the header
                    if len(row_data) == len(header_row):
                        table_data.append(row_data)

            # Convert the data to a DataFrame
            df = pd.DataFrame(table_data, columns=header_row)
            dfs.append(df)

            # Save the DataFrame to a CSV file (you can choose another format if needed)
            df.to_csv(f'table_{i + 1}.csv', index=False)

        # Return the list of DataFrames
        return dfs

    else:
        print(f'Error: Unable to fetch data from {url}. Status code: {response.status_code}')
        return None

# Example usage:
url_to_fetch = 'https://investors.canoo.com/financial-information/cash-flow'
dataframes = retrieve_tables_to_dataframe(url_to_fetch)

# Access individual DataFrames in the list if needed
if dataframes:
    for i, df in enumerate(dataframes):
        print(f'DataFrame for Table {i + 1}:\n{df}\n')
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

def retrieve_tables_to_dataframe(url, excel_file):
    # Send a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find all tables in the HTML
        tables = soup.find_all('table')

        # Create an Excel writer object
        with pd.ExcelWriter(excel_file) as writer:
            # Iterate through each table and convert it to a DataFrame
            for i, table in enumerate(tables):
                # Extract data from the table
                table_data = []
                header_row = None

                for row in table.find_all('tr'):
                    row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]

                    # If header_row is not set, assume the current row is the header
                    if header_row is None:
                        header_row = row_data
                    else:
                        # Check if the current row has the same number of columns as the header
                        if len(row_data) == len(header_row):
                            table_data.append(row_data)

                # Convert the data to a DataFrame
                df = pd.DataFrame(table_data, columns=header_row)

                # Write the DataFrame to the Excel file as a sheet
                df.to_excel(writer, sheet_name=f'Table_{i + 1}', index=False)

        print(f'DataFrames saved to {excel_file}')

    else:
        print(f'Error: Unable to fetch data from {url}. Status code: {response.status_code}')

# Example usage:
url_to_fetch = 'https://investors.canoo.com/financial-information/cash-flow'
excel_file = 'tables_data.xlsx'
retrieve_tables_to_dataframe(url_to_fetch, excel_file)
```
